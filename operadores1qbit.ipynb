

## Operadores de rotação

$$R_x(\theta) = 
\begin{bmatrix}
cos\left(\theta/2\right) & -i sen(\theta/2) \\
-i sen(\theta/2) & cos\left(\theta/2\right)
\end{bmatrix} 
%
R_y(\theta) = 
\begin{bmatrix}
cos\left(\theta/2\right) & -sen(\theta/2) \\
sen(\theta/2) & cos\left(\theta/2\right)
\end{bmatrix} 
%
R_x(\theta) = 
\begin{bmatrix}
e^{-i\theta/2} & 0 \\
0 & e^{i\theta/2}
\end{bmatrix} 
$$
### Esfera de Bloch


```python
import qutip # http://qutip.org/
import numpy as np # https://numpy.org/
```

```python
qutip.snot() #Hadamard
```

```python
# H|0> na esfera de Bloch
b = qutip.Bloch() # gera um esfera de Bloch
state = qutip.basis(2,0) # |0>
b.add_states(state) # Adiciona um estado na esfera de bloch
state = qutip.snot() * state
b.add_states(state) 
b.show() # Imprime a esfera
```

```python
# H|1> na esfera de Bloch
b = qutip.Bloch() 
state = qutip.basis(2,1) # |1>
b.add_states(state) 
state = qutip.snot() * state
b.add_states(state) 
b.show()
```

```python
state = qutip.basis(2,0)
state = qutip.rx(np.pi/2) * state # (|0> -i|1>)/sqrt(2)
b = qutip.Bloch() 
b.add_states(state) 
b.show()
```

```python
state = qutip.basis(2,0)
state = qutip.rx(-np.pi/2) * state # (|0> + i|1>)/sqrt(2)
b = qutip.Bloch() 
b.add_states(state) 
b.show()
```

### Ação dos operadores $R_x$, $R_y$, $R_z$ na esfera de Bloch
#### Visualização R_x

```python
# visualização |0> na esfera de Bloch
state = qutip.basis(2,0) # Inicializa state com |0>
b = qutip.Bloch() # gera um esfera de Bloch
b.add_states(state) # Adiciona um vetor representado state na esfera
b.show() # Imprime a esfera
```

Ao aplicar $R_x(\theta)$ rotacionamos o vetor

```python
state = qutip.rx(np.pi/10) * state
b.add_states(state)
b.show()
```

```python
for k in range(10):
    state = qutip.rx(np.pi/10) * state
    b.add_states(state)
b.show()
```

#### Visualização $R_y$

```python
# visualização |0> na esfera de Bloch
state = qutip.basis(2,0) # Inicializa state com |0>
b = qutip.Bloch() # gera um esfera de Bloch
b.add_states(state) # Adiciona um vetor representado state na esfera
b.show() # Imprime a esfera
```

```python
state = qutip.ry(np.pi/10) * state
b.add_states(state)
b.show()
```

```python
for k in range(10):
    state = qutip.ry(np.pi/10) * state
    b.add_states(state)
b.show()
```

#### Visualização $R_z$

```python
# visualização |+> na esfera de Bloch
state = qutip.snot() * qutip.basis(2,0) # Inicializa state  R_x(\pi/2) * |0>
b = qutip.Bloch() # gera um esfera de Bloch
b.add_states(state) # Adiciona um vetor representado state na esfera
b.show() # Imprime a esfera
```

```python
state = qutip.rz(np.pi/10) * state
b.add_states(state)
b.show()
```

```python
for k in range(10):
    state = qutip.rz(np.pi/10) * state
    b.add_states(state)
b.show()
```

1. Determine $\alpha$, $\beta$, $\gamma$ e $\delta$ para o operador X.

```python
alpha = np.pi/2
beta = 0
gamma = np.pi
delta = np.pi
np.e**(1j * alpha) * qutip.rz(beta) * qutip.ry(gamma) * qutip.rz(delta)
```
